```python
import sys
sys.path.insert(0, '..')
import warnings; warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data.loader import DataLoader, load_config
from src.visualization.plots import Plotter

cfg = load_config('../configs/params.yaml')
loader = DataLoader('../configs/params.yaml')
plotter = Plotter(cfg)
print("Modules loaded OK")
```

## 1. Tải và kiểm tra dữ liệu gốc

```python
df = loader.load_raw()
loader.describe(df)
```

## 2. Data Dictionary
| Cột | Ý nghĩa |
|-----|---------|
| age | Tuổi |
| sex | Giới tính (0=Nữ,1=Nam) |
| cp | Loại đau ngực (0-3) |
| trestbps | Huyết áp nghỉ (mmHg) |
| chol | Cholesterol (mg/dl) |
| fbs | Đường huyết đói >120 |
| restecg | ECG nghỉ ngơi |
| thalach | Nhịp tim tối đa |
| exang | Đau ngực khi gắng sức |
| oldpeak | ST depression |
| slope | Độ dốc ST peak |
| ca | Số mạch máu lớn |
| thal | Thalassemia |
| **target** | **Bệnh tim (0/1)** |

## 3. Phân phối biến mục tiêu

```python
plotter.plot_target_distribution(df['target'])
print(f"Tỷ lệ bệnh tim: {df['target'].mean()*100:.1f}%")
print(f"Imbalance ratio: {df['target'].value_counts()[0]/df['target'].value_counts()[1]:.2f}:1")
```

## 4. Phân phối các đặc trưng số

```python
num_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
plotter.plot_numerical_distributions(df, num_cols)
df[num_cols].describe().round(2)
```

## 5. So sánh phân phối theo nhãn bệnh tim

```python
plotter.plot_boxplots_by_target(df, num_cols)
```

## 6. Đặc trưng phân loại theo target

```python
cat_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
plotter.plot_categorical_counts(df, cat_cols)
```

## 7. Ma trận tương quan

```python
plotter.plot_correlation_heatmap(df)
corr_with_target = df.corr()['target'].drop('target').sort_values(key=abs, ascending=False)
print("Tương quan với target:")
print(corr_with_target.round(3))
```

## 8. Missing values và bất thường

```python
plotter.plot_missing_values(df)
# Kiểm tra giá trị bất thường
print("Giá trị 0 trong các cột số:")
for col in ['chol', 'trestbps']:
    print(f"  {col}==0: {(df[col]==0).sum()} bản ghi")
print(f"  ca==4: {(df['ca']==4).sum()} bản ghi")
print(f"  thal==0: {(df['thal']==0).sum()} bản ghi")
```
